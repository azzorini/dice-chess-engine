```python
import chess
```

```python
from DiceChess import DiceBoard, pieces_list, pieces_name_list
```

```python
from IPython.display import display
```

```python
### Start game
board: DiceBoard = DiceBoard()
is_game_over: bool = False
while not is_game_over:
    board.roll_dices()
    player_str: str = "White" if board.turn == chess.WHITE else "Black"
    print("(" + player_str +") Roll: ", [pieces_name_list[p-1] for p in board.dice_roll])
    move_list: list = board.get_moves()
    nMoves: int = len(move_list)
    while nMoves > 0 and not is_game_over:
        print("(" + player_str + ") Remaining pieces: ", [pieces_name_list[p-1] for p in board.dice_roll])
        display(board)
        print("Available moves:")
        for i, move in enumerate(move_list, start=1):
            print('\t', i, board.san(move))
        
        choice: int = 0
        while choice < 1 or choice > nMoves:
            print(f"Choose a move (1-{nMoves}): ")
            try:
                choice = int(input())
            except ValueError:
                choice = 0
        
        is_game_over = board.is_game_over(move_list[choice-1])
        
        board.push(move_list[choice-1], make_changes = True)
        
        move_list = board.get_moves()
        nMoves = len(move_list)
    
    if is_game_over:
        print(player_str + " wins.")
    else:
        print("Next player (current player out of moves)")
        board.next_player()
    display(board)
```
